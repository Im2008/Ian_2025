## <span class ="glow">Main strings hack (20%)</span>
Using sections above...
1. Make a text analyzer with strings that accepts input from the user
    - Takes multiple words/ sentences
    - Displays original string
    - Counts total characters (includes spaces)
    - Displays longest word with how many characters it has
    - Displays the string reversed
    - Find middle word/ character (excludes spaces or special characters)
    - Correctly displays everything!
<br>

2. Other criteria:
    - make sure program can process upper/lowercase letters, numbers, and symbols
    - Customize and add your own unique function
    - Try to use both python and js, either making two separate analyzers or using both at the same time

3. Extra/ Bonus: make two additional input systems where user can replace words of the original input. 
    - The first input is which word they want to replace, the second is what to replace it with
    - For example, if they inputed a song verse, through the second input they can replace a specific word that repeats in the song with another
    - Make sure input can handle requests to change things that aren't in original string.


```python
#Popcorn Hack 7

# Example song lyrics (clean)
lyrics = """
This is the first verse, first verse in the song
And this is the chorus, chorus of the song
This is the second verse, verse two of the song
And this is the last chorus, chorus of the song
"""

# 1. Find how many times the title of the song (in this case, "song") appears in the lyrics
def count_title_occurrences(lyrics, title):
    title_count = lyrics.lower().count(title.lower())  # Case insensitive count
    print(f"The word '{title}' appears {title_count} times in the lyrics.")
    return title_count

# 2. Find the index of the 50th word in the lyrics
def find_50th_word_index(lyrics):
    words = lyrics.split()  # Split the lyrics into a list of words
    if len(words) >= 50:
        index_50th_word = lyrics.find(words[49])  # Find the index of the 50th word
        print(f"The 50th word is '{words[49]}' and its index is {index_50th_word}.")
        return index_50th_word
    else:
        print("The lyrics have less than 50 words.")
        return -1

# 3. Replace the first verse with the last verse
def replace_first_with_last(lyrics):
    verses = lyrics.split("\n")
    first_verse = verses[1]  # Assume second line is the first verse
    last_verse = verses[-2]  # Assume second to last line is the last verse
    print(f"\nOriginal first verse: {first_verse}")
    print(f"Last verse: {last_verse}")
    
    # Replace the first verse with the last verse
    verses[1] = last_verse
    updated_lyrics = "\n".join(verses)
    print("\nUpdated lyrics with first verse replaced by the last verse:")
    print(updated_lyrics)
    return updated_lyrics

# 4. Concatenate two verses to create a new song
def concat_two_verses(lyrics):
    verses = lyrics.split("\n")
    verse_1 = verses[1]  # Assume second line is the first verse
    verse_3 = verses[3]  # Assume fourth line is the second verse
    new_song = verse_1 + " " + verse_3  # Concatenate two verses
    print("\nNew song created by concatenating two verses:")
    print(new_song)
    return new_song

# Main execution
title = "song"

# 1. Count the occurrences of the song title in the lyrics
count_title_occurrences(lyrics, title)

# 2. Find the index of the 50th word
find_50th_word_index(lyrics)

# 3. Replace the first verse with the last verse
replace_first_with_last(lyrics)

# 4. Concatenate two verses to make your own song
concat_two_verses(lyrics)
```

```python
// HOMEWORK HACK

// Function to get user input and analyze the text
function analyzeText() {
    let userInput = prompt("Enter multiple words or a sentence:");
    
    // Display the original string
    console.log("Original string: ", userInput);

    // 1. Counting total characters (including spaces)
    let totalCharacters = userInput.length;
    console.log("Total characters (including spaces):", totalCharacters);

    // 2. Finding the longest word
    let words = userInput.split(" ");  // Split into an array of words
    let longestWord = "";
    
    words.forEach(word => {
        if (word.length > longestWord.length) {
            longestWord = word;
        }
    });
    
    console.log("Longest word: ", longestWord, "with", longestWord.length, "characters");

    // 3. Display the string reversed
    let reversedString = userInput.split("").reverse().join("");
    console.log("Reversed string: ", reversedString);

    // 4. Finding the middle word (ignoring spaces and special characters)
    let filteredInput = userInput.replace(/[^a-zA-Z0-9]/g, "");  // Removing spaces and special characters
    let middleIndex = Math.floor(filteredInput.length / 2);
    let middleCharacter = filteredInput[middleIndex];
    console.log("Middle character: ", middleCharacter);

    // 5. Additional unique function: Count the number of vowels in the string
    let vowelCount = userInput.match(/[aeiouAEIOU]/g) ? userInput.match(/[aeiouAEIOU]/g).length : 0;
    console.log("Total vowels: ", vowelCount);
}

// Call the text analyzer function
analyzeText();


// 2. Bonus: Additional input systems to replace words

function replaceWords() {
    let originalInput = prompt("Enter the original string:");
    let wordToReplace = prompt("Which word would you like to replace?");
    let replacementWord = prompt("What would you like to replace it with?");

    // Replacing the word in the original input
    let replacedString = originalInput.replace(new RegExp(wordToReplace, 'g'), replacementWord);

    // Check if the word was found and replaced
    if (originalInput === replacedString) {
        console.log(`The word "${wordToReplace}" was not found in the original string.`);
    } else {
        console.log("Original string: ", originalInput);
        console.log(`Modified string: `, replacedString);
    }
}

// Call the replace words function
replaceWords();

```
